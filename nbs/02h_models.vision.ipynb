```python
#| export
def create_model(
    model_name: str,
    pretrained: bool = False,
    num_classes: int = 10,
    **kwargs
) -> nn.Module:
    """
    Create a vision model.
    
    Args:
        model_name: Model architecture and variant separated by underscore (e.g., 'resnet_50', 'vit_base')
        pretrained: Whether to load pretrained weights
        num_classes: Number of output classes
        **kwargs: Additional model-specific arguments
    
    Returns:
        Initialized model
    
    Examples:
        >>> model = create_model('resnet_50', num_classes=1000)
        >>> model = create_model('vit_base', num_classes=10, pretrained=True)
    """
    architecture, variant = parse_model_name(model_name)
    factory = get_factory(architecture)
    
    return factory(
        variant=variant,
        pretrained=pretrained,
        num_classes=num_classes,
        **kwargs
    )
```

```python
#| hide
import random, torch
import numpy as np
def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
seed_everything()
model_1 = create_model(model_name='efficientnet_b0', pretrained=False, num_classes=10)
seed_everything()
model_2 = create_model(model_name='efficientnet_b0', pretrained=False, num_classes=10)
```

```python
#| hide
# check if models are the same instance (should be False)
for p1, p2 in zip(model_1.parameters(), model_2.parameters()):
    if not p1.equal(p2):
        print("no")
```
