```python
#| hide
import pandas as pd
df = pd.read_csv("wandb_export_2025-03-25T09_51_30.969+02_00.csv")
df.head()

df2 = pd.read_csv("wandb_export_2025-03-25T09_57_01.074+02_00.csv")
```

```python
y1 = df['example_project20250325_014854 - test_accuracy_score']
x1 = df["Step"]

y2 = df2['example_project20250320_151241 - test_accuracy_score']
x2 = df2["Step"]    
# plot the data
import matplotlib.pyplot as plt
plt.plot(x1, y1, label='FMTL-Graph')
plt.plot(x2, y2, label='Fedu')
plt.title("Accuracy score")
plt.xlabel("Step")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.savefig("accuracy.png")
plt.show()
```

```python
df3 = pd.read_csv("wandb_export_2025-03-25T10_03_21.702+02_00.csv")
y = df3['example_project20250320_151241 - test_loss']
x = df3["Step"]
df4 = pd.read_csv("wandb_export_2025-03-25T10_03_58.304+02_00.csv")
y2 = df4['example_project20250325_014854 - test_loss']
x2 = df4["Step"]
plt.plot(x2, y2, label='FMTL-Graph')
plt.plot(x, y, label='Fedu')
plt.title("Test Loss")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.grid()
plt.legend()
# save the plot
plt.savefig('test_loss.png')
plt.show()
```
