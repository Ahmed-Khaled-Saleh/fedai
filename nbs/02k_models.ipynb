```python
#| export
import random
import numpy as np
import os
import torch

import fedai.models.vision as vision_models
```

```python
#| export
def create_model(cfg):

    torch.manual_seed(cfg.random_seed)  
    np.random.seed(cfg.random_seed)
    random.seed(cfg.random_seed)
    
    if cfg.model.model_name in vision_models.list_model_names() and cfg.data.modality == ['Vision']:
        return vision_models.create_model(**cfg.model)
    
    raise ValueError(f"Model '{cfg.model.model_name}' is not recognized, the available models are: {vision_models.list_model_names()}")
```

```python
#| hide
from omegaconf import OmegaConf
```

```python
#| hide
cfg = OmegaConf.create({
    "model": {
        "model_name": "lenet_fedavg",
        "pretrained": False,
        "num_classes": 20,
        "img_size" : (3, 32, 32),
        "hidden_dim": 256
    },
    "random_seed": 42,
    "data":{
        "modality": ['Vision']
    }
})
model = create_model(cfg)
```
