```python
#| export 
from fastcore.utils import *
from dataclasses import dataclass, field
from omegaconf import OmegaConf, MISSING
from typing import List, Optional, Tuple, Dict, Any, Union, Literal, Any
```

```python
#| export

@dataclass
class BaseOptimizerConfig:
    cls: str = "SGD"
    lr: float = 0.01
   


@dataclass
class SGDConfig(BaseOptimizerConfig):
    momentum: float = 0.5
    nesterov: bool = False

@dataclass
class MuonConfig(SGDConfig):
    cls: str = "Muon"
    lr: float = 0.01
    weight_decay: float = 5e-4
    ns_coefficients: Tuple[float, float, float] = (3.4445, -4.775, 2.0315)


@dataclass
class AdamConfig(BaseOptimizerConfig):
    cls: str = "Adam"
    lr: float = 0.001
    betas: Tuple[float, float] = (0.9, 0.999)
    eps: float = 1e-08
    amsgrad: bool = False
    weight_decay: float = 0.0


@dataclass
class AdamWConfig(AdamConfig):
    cls: str = "AdamW"


@dataclass
class PerFedavgOptimizerConfig(BaseOptimizerConfig):
    cls: str = "PerFedAvgOpt"

@dataclass
class pFedMeOptimizerConfig(BaseOptimizerConfig):
    cls: str = "pFedMeOptimizer"
    lambda_: float = 10
    mu: float = 0.0

@dataclass
class FedProxOptimizerConfig(BaseOptimizerConfig):
    cls: str = "PerturbedGradientDescent"
    mu: float = 0.0

@dataclass
class APFLOptimizerConfig(BaseOptimizerConfig):
    cls: str = "APFLOptimizer"
```

```python
#| hide
from fedai.models import create_model
from fedai.optimizers import get_optimizer
from omegaconf import OmegaConf
from fedai.client_selector import BaseClientSelector
import importlib
from fedai.cfgs.data import *
from fedai.cfgs.models import *
from fedai.cfgs.algos import *
```

```python
#| hide
adam_cfg = BaseOptimizerConfig(cls="Adam", lr=0.001)
cfg = OmegaConf.create({'model': LeNetConfig(), 'data': MNISTConfig(), 'algorithm': FedAvgConfig(), 'optimizer': adam_cfg})
model = create_model(cfg)
```

```python
#| hide
opt_cls = getattr(importlib.import_module("torch.optim"), adam_cfg.cls)
kwargs = {k: v for k, v in vars(adam_cfg).items() if k != "cls"}
optimizer = opt_cls(model.parameters(), **kwargs)
```

```python
#| hide
import torch.optim as optim
getattr(optim, adam_cfg.cls)(model.parameters(), **kwargs)
```
