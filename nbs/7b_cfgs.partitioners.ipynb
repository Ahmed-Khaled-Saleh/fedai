```python
#| export 
from fastcore.utils import *
from dataclasses import dataclass, field
from typing import List, Optional, Tuple, Dict, Any, Union, Literal
from enum import Enum
import numpy as np
from flwr_datasets.partitioner import *
from fedai.data.vision.partitioners import *   
```

```python
#| export
@dataclass
class PartitionerConfig:
    cls: Optional[str] = None
    
@dataclass
class PathologicalConfig(PartitionerConfig):
    cls: str = "PathologicalPartitioner"
    num_partitions: int = 20
    partition_by: str = 'label'
    num_classes_per_partition: int = 2
    class_assignment_mode: str = "deterministic"# "deterministic" # random
    
@dataclass
class DirichletConfig(PartitionerConfig):
    cls: str = "DirichletPartitioner"
    num_partitions: int = 20
    partition_by: str = 'label'
    alpha: float = 0.5

@dataclass
class SizePartitionerConfig(PartitionerConfig):
    cls: str = "SizePartitioner"
    partition_sizes: List[int] = field(default_factory=lambda: [1500, 2000])

@dataclass
class RotatedBatchedConfig(PartitionerConfig):
    cls: str = "RotatedPatchedPartitioner"
    num_partitions: int = 20
    num_rotations: int = 4
    patch_size: int = 2

@dataclass
class VariableClassPartitionerConfig(PartitionerConfig):
    cls: str = "VariableClassPartitioner"
    num_partitions: int = 20
    partition_by: str = 'label'
    num_classes_per_partition: Dict[int, int] = field(default_factory=lambda: {i: 2 for i in range(20)}) # default 2 classes per partition
    default_num_classes: int = 2


@dataclass
class GroupedClassPartitionerConfig(PartitionerConfig):
    cls: str = "GroupedClassPartitioner"
    num_partitions: int = 20
    partition_by: str = 'label'
    num_groups: int = 4
    num_classes_per_group: int = 3
    num_classes_per_client: int = 2


@dataclass
class GroupedDistributionPartitionerConfig(PartitionerConfig):
    cls: str = "GroupedDistributionPartitioner"
    num_partitions: int = 20
    partition_by: str = 'label'
    num_groups: int = 4
    num_classes_per_group: int = 3
    num_classes_per_client: int = 2
    distribution_array: List[List[float]] = field(default_factory=lambda: np.random.lognormal(0, 1, size=(2, 20)).tolist())   # default random log-normal distribution for 2 class-slots × 20 clients
```

```python
#| hide
import importlib
cfg = PathologicalConfig()
args = {k: v for k, v in cfg.__dict__.items() if k != 'cls'}
cls = getattr(importlib.import_module("flwr_datasets.partitioner"), cfg.cls)
p = cls(**args)


cfg = DirichletConfig()
args = {k: v for k, v in cfg.__dict__.items() if k != 'cls'}
cls = getattr(importlib.import_module("flwr_datasets.partitioner"), cfg.cls)
p = cls(**args)


cfg = SizePartitionerConfig()
args = {k: v for k, v in cfg.__dict__.items() if k != 'cls'}
cls = getattr(importlib.import_module("flwr_datasets.partitioner"), cfg.cls)
p = cls(**args)


cfg = RotatedBatchedConfig()
args = {k: v for k, v in cfg.__dict__.items() if k != 'cls'}
cls = getattr(importlib.import_module("fedai.data.vision.partitioners"), cfg.cls)
p = cls(**args)

cfg = VariableClassPartitionerConfig()
args = {k: v for k, v in cfg.__dict__.items() if k != 'cls'}
cls = getattr(importlib.import_module("fedai.data.vision.partitioners"), cfg.cls)
p = cls(**args)


cfg = GroupedClassPartitionerConfig()
args = {k: v for k, v in cfg.__dict__.items() if k != 'cls'}
cls = getattr(importlib.import_module("fedai.data.vision.partitioners"), cfg.cls)
p = cls(**args)

cfg = GroupedDistributionPartitionerConfig()
args = {k: v for k, v in cfg.__dict__.items() if k != 'cls'}
cls = getattr(importlib.import_module("fedai.data.vision.partitioners"), cfg.cls)
p = cls(**args) 
p
```
