```python
#| export 
from fastcore.utils import *
import pandas as pd
import wandb
import os
import argparse
import numpy as np
from dataclasses import dataclass, field
from omegaconf import OmegaConf, MISSING
from typing import List, Optional, Tuple, Dict, Any, Union, Literal, Any
from fedai.cfgs.models import *
from fedai.cfgs.data import *
from fedai.cfgs.algos import *
from fedai.cfgs.gpu_servers import *
```

```python
#| export
@dataclass
class MainConfig:

    defaults: List[Any] = field(default_factory=lambda: [
        "_self_",               # Use values in this class
        {"model": "lenet"},     # Default group choice
        {"data": "mnist"},    # Default group choice
        {"partitioner": "pathological"},    # Default group choice
        {"optimizer": "sgd"},    # Default group choice
        {"algorithm": "fedavg"},
        {"server": "puhti"}
    ])


    data: Any = MISSING
    partitioner: Any = MISSING
    model: Any = MISSING 
    optimizer: Any = MISSING 
    algorithm: Any = MISSING 
    server: Any = MISSING 
    
    num_clients: int = 20
    m: float = 1.0
    n_rounds: int = 200 
    local_epochs: int = 5
    random_seed: int = 42
    batch_size: int = 128
    now: Optional[str] = None

    project_name: str = "$Fedas_test_${data.name}"#"${algorithm.name}_${data.name}_${model.name}_"

    writer: str = "WandbWriter"

    training_metrics: List[str] = field(default_factory=lambda: ["accuracy_score"])
    test_metrics: List[str] = field(default_factory=lambda: ["accuracy_score"])
```

```python
#| hide
cfg = MainConfig()
cfg.algorithm, cfg.model, cfg.data
cfg.training_metrics
```
